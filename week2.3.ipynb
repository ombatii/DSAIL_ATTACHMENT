# linear regression on Salary Dataset 

## 1. Business understanding
Verify this equation using Salary dataset;
 y = mx + c

## Import libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import statsmodels.api as sm
```

## Import dataset

```python
df = pd.read_csv("Salary_dataset.csv")
df.head()
```

## Data analysis

```python
df.info()
```

```python
df.isna().sum()
```

There are no missing data

```python
"""
Let us say our y is Salary and x is YearsExperience.We can fit the above equation using lmplot() function from seaborn.

"""

## scatterplot
sns.lmplot(x="YearsExperience", y="Salary", data=df);
```

## Perform Linear Regression 

```python
# define response variable
y = df['Salary']

# define explanatory variable
x = df[['YearsExperience']]

# add constant to predictor variables
x = sm.add_constant(x)

# fit linear regression model
model = sm.OLS(y, x).fit()

# our model summary
print(model.summary())
```

so our c = 2.485e+04 and m = 9449.9623 ,hence our linear equation is y = 9449.9623 x + 2.485e+04

## Residual plots

### 1.Residual vs. fitted values plot

```python
# define figure size
fig = plt.figure(figsize=(12,8))

# produce residual plots
fig = sm.graphics.plot_regress_exog(model, 'YearsExperience', fig=fig)
```

From the residual plots we can get this insight;
From the one in the top right corner is the residual vs. fitted plot; The residuals appear to be randomly and evenly distributed throughout the chart around the value zero hence we can say that homoscedasticity is not violated.
  > In linear regression, "homoscedastic" refers to the assumption of equal variance of the errors or residuals across all levels of the predictor variables.
